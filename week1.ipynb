```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt

steps = pd.read_csv('steps.csv', sep=';')
steps.head()
```

```python
survey = pd.read_csv('survey.csv')
survey.head()
```

```python
df = pd.merge(steps, survey, on = 'id')
df.head()
```

```python
df['AverageSteps'] = df.mean(numeric_only=True, axis=1) # Average steps per every id
df.head()
```

```python
df.info() # Info about variables
```

```python
df.fillna(0) # Getting rid of NA - replacing with 0
```

```python
#Remove data with < 60000 steps 

def above_60000(steps):
    if(steps > 60000.0): 
        return float('NaN')
    else: 
        return steps
df['AverageSteps'] = df['AverageSteps'].apply(above_60000)

df.head()
```

```python
#Graph 1 

df['AverageSteps'].name = 'Average steps'
Average_steps = df['AverageSteps'].dropna() #Dropping NaN's
sns.distplot(Average_steps, kde=False) 
plt.title('Distribution - average steps')
plt.show()
```

```python
#Graph 2 Average of steps daily
sns.boxplot(df.AverageSteps)
plt.title('Average of steps daily')
plt.xlabel('Steps')
plt.show()
```

```python
#Graph 3 Distribution mean - Steps per day
sns.kdeplot(df.AverageSteps)
```

```python
#Summary of statistics

mean = df['AverageSteps'].mean()
median = df['AverageSteps'].median()
variance = df['AverageSteps'].var()

print(f'The median daily steps is: {median}')
print(f'The mean daily steps is: {mean}')
print(f'The variance daily steps is: {variance}')

# I do not know why but print(f'') gives me an Syntax error.
```
